```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
data = iris.data
label = iris.target
columns = iris.feature_names
```

```python
data = pd.DataFrame(data, columns=columns)
data.head()
```

```python
data.shape
```

```python
# 데이터의 준비
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, label, test_size = 0.2, random_state = 2022 )

# Logistic Regression : 회귀같지만, 회귀같지않은 !
# - 직선을가지고, 분류를 하는것
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)  # 학습
y_pred = lr.predict(X_test)  # 예측
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('로지스틱 회귀, 정확도: {:.2f}'.format(accuracy_score(y_test, y_pred)))  # 테스트용데이터의 실제값, 테스트한값을 비교하겠다
```

```python
# 서포트벡터

from sklearn.svm import SVC
svc = SVC(C=100)   # C의 값은 데이터에 따라 다르게 해주어야함
```

```python
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
print('서포트 벡터머신, 정확도: {:.2f}'.format(accuracy_score(y_test, y_pred))) #정확도확인을위해 결과값 찍어보기
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth =5)

dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)
print('결정 트리, 정확도 {:.2f}'.format(accuracy_score(y_test, y_pred)))
```

```python
# Random Forest, 여러개의 Tree를 이용하여 의사결정
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth = 5)

rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
print('랜덤 포레스트, 정확도 {:2f}'.format(accuracy_score(y_test, y_pred)))
```
