```python
from keras.datasets import mnist
```

```python
(train_images, train_labels),(test_images, test_labels) = mnist.load_data()
```

```python
train_images.shape    # 6만개의 데이터, 실제한장의 사진크기는 28*28
```

```python
len(train_labels)  # 라벨데이티가 6만개 준비되어있음
```

```python
test_images.shape
```

```python
# 케라스에서 모델과 레이어 불러옴
from keras import models
from keras import layers

# 어떤모델을 쓸건지는 신경망에 따라 결정됨

network = models.Sequential() # Sequential : 순차적으로 신경망을 하나씩 만들것이다

# 입력층, 출력층 두가지 레이어로 구성
# 입력층, 일반적인 layer = dense layer
network.add(layers.Dense(512, activation='relu', input_shape = (28 * 28,)))
# 출력층 : 출력은 0에서 9까지 10개, 여러개 중에서 하나가 나오는 softmax
network.add(layers.Dense(10, activation = 'softmax'))     
```

```python
# compile 시킬때 3가지 옵션을 줘야함
network.compile(optimizer='rmsprop',  # 최적화 => optimizer는 가중치와 bias값을 조절
                loss='categorical_crossentropy', # 다중분류 문제에서는 categorical_crossentropy를 보통 사용     
                                                 # 이진분류문제에선 binary crossentropy
                metrics=['accuracy'])    # 정확도 기준 평가
```

```python
# 데이터 타입의 변환
# (60000, (28, 28)) 이 데이터를  --> ( 60000, 784 )로 차원 줄여서 사용한다(컴퓨터가 이해를 더 쉽게한다)

train_images = train_images.reshape((60000, 28*28))
train_images = train_images.astype('float32') / 255  # 0붙터 255까지의값을 255로 나눠서 보기 편하게 다 소수점임

# 0번 데이터 출력
# print(train_images[0])

# 테스트용 데이터
test_images = test_images.reshape((10000, 28*28))
test_images = test_images.astype('float32')  / 255
```

```python
# 분류형 데이터 설정, 학습데이터가 아니라 분류데이터로 인코딩
from tensorflow.keras.utils import to_categorical

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
network.fit(train_images, train_labels, epochs=5, batch_size=128)   # epochs : 반복횟수, batch_size : 한번에 처리하는 데이터 수
# loss = 실제값과 차이
```

```python
# 평가
test_loss, test_acc = network.evaluate(test_images, test_labels)
```

```python
print('test_acc: ', test_acc)   # 97.7%
```
